```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
class LinearRegression:

    def __init__(self,lr=0.001,niters=1000):
        self.lr = lr
        self.niters=niters
        self.weights= None
        self.bias=None

    def fit(self,x,y):
        nsamples,nfeatures=x.shape
        self.weights = np.zeros(nfeatures)
        self.bias = 0

        for _ in range(self.niters):
            ypred=np.dot(x,self.weights)+self.bias
            dw=(1/nsamples)*np.dot(x.T,(ypred-y))
            db=(1/nsamples)*np.sum(ypred-y)
            self.weights = self.weights - self.lr * dw
            self.bias = self.weights - self.lr * dw

    def predict(self,x):
        ypred = np.dot(x,self.weights)+self.bias
        return ypred
        
```

```python
X,y = datasets.make_regression(n_samples=100,n_features =1, noise = 20, random_state=4)
xtrain,xtest,ytrain,ytest = train_test_split(X,y,test_size=0.2, random_state=1234)
```

```python
fig=plt.figure(figsize=(8,6))
plt.scatter(X[:,0],y,color="b", marker ="o", s =30)
plt.show()
```

```python
reg=LinearRegression()
reg.fit(xtrain,ytrain)
predictions = reg.predict(xtest)
```

```python
def mse(ytest,predictions):
    return np.mean((ytest-predictions)**2)
```

```python
mse= mse(ytest,predictions)
```
